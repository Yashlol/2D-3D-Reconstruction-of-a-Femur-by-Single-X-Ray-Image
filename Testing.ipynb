```python
import json
import os

folder = r"C:\Users\Yash\OneDrive\Desktop\Nit Internship\Base paper Implementation\Dataset\Experimental Data"

files = [
    "Center of femoral head_04.mrk.json",
    "Most inferior point of head_04.mrk.json",
    "Most superior point of head_04.mrk.json",
    "Tip of greater trochanter_04.mrk.json",
    "Lateral-most Tip_04.mrk.json",
    "Tip of lesser trochanter_04.mrk.json",
    "Posterior Shaft Edge_04.mrk.json",
    "Anterior Shaft Edge_04.mrk.json",
    "Lateral Shaft Edge_04.mrk.json",
    "Shaft Center at Cut Plane_04.mrk.json",
    "Neck center_04.mrk.json"
]

landmarks = {}

for file in files:
    filepath = os.path.join(folder, file)

    with open(filepath, "r") as f:
        data = json.load(f)

    point = data["markups"][0]["controlPoints"][0]["position"]

    name = os.path.basename(file).replace(".mrk.json", "")

    landmarks[name] = point

output_file = os.path.join(folder, "femur_04_landmarks.json")

with open(output_file, "w") as f:
    json.dump({"landmarks": landmarks}, f, indent=4)

print(f"Merged successfully: {output_file}")
```
